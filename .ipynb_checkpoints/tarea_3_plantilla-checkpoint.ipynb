4. Count the number of duplicates in the Google Play dataset. You must build a function to loop through the Google Play dataset

````python
# Solution
df_google.duplicated().sum()
````

````python
def count_duplicates(dataframe):
    num_duplicates = 0
    seen_rows = set()

    for index, row in dataframe.iterrows():
        #convierte la fila a una tupla y comprueba si ya la hemos visto
        row_tuple = tuple(row)
        if row_tuple in seen_rows:
            num_duplicates += 1
        else:
            seen_rows.add(row_tuple)
    return num_duplicates

count_duplicates(df_google)
````

6. Remove non-english apps from both Android and Apple datasets. You should write a function that takes a string, if there's more than three characters that fall outside the ASCII range (0 - 127), then the function should return False (identify the string as non-English), otherwise it should return True. Hint: use the [```ord()```  function](https://www.w3schools.com/python/ref_func_ord.asp)

````python
def is_english(string):
    non_ascii_count = 0
    for char in string:
        if ord(char) > 127:
            non_ascii_count += 1
        if non_ascii_count > 3:
            return False
    return True
````

````python
# Filtrar aplicando la función a las columnas relevantes en ambos conjuntos de datos
google_en = df_google[df_google['App'].apply(is_english)]
apple_en = df_apple[df_apple['track_name'].apply(is_english)]

# google_en, apple_en
````

 8. Build a frequency table for the ```prime_genre``` column of the App Store dataset, and the ```Genres``` and ```Category``` columns of the Google Play data set. What are the most common genres for each market? You must develop and deploy your function

````python
# Solution

def build_frequency_table(column, data):
    frequency_table = {}
    for value in data[column]:
        if value in frequency_table:
            frequency_table[value] += 1
        else:
            frequency_table[value] = 1
    return frequency_table

# Para la App Store
apple_genre_frequency = build_frequency_table('prime_genre', apple_english)
print("App Store - Prime Genre Frequency:")
print(apple_genre_frequency)

# Para Google Play - Genres
google_play_genres_frequency = build_frequency_table('Genres', android_english)
print("\nGoogle Play - Genres Frequency:")
print(google_play_genres_frequency)

# Para Google Play - Category
google_play_category_frequency = build_frequency_table('Category', android_english)
print("\nGoogle Play - Category Frequency:")
print(google_play_category_frequency)
````

 9. One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre. Based on that exercise, give one app profile recommendation for Google Play.
 
    For the Google Play data set, we can find the required information in the ```Installs``` column. Then, compute the average number of installs by genre. Hint 1: consider only unique values of the ```Category``` column. Hint 2: If an app has, for instance, ```1,000,000+``` assume it as ```1,000,000``` installs. Hint 3: Be careful with a skewed distribution

````python
# Solution

# Asumiendo que 'android_english' es el conjunto de datos limpio de Google Play
# Asegúrate de que la columna 'Installs' sea numérica y libre de caracteres no numéricos
android_english['Installs'] = android_english['Installs'].str.replace(',', '')
android_english['Installs'] = android_english['Installs'].str.replace('+', '')
android_english['Installs'] = android_english['Installs'].astype(int)

# Agrupar por género y calcular el promedio de instalaciones
genre_avg_installs = android_english.groupby('Category')['Installs'].mean()

# Ordenar por instalaciones promedio en orden descendente
sorted_genres = genre_avg_installs.sort_values(ascending=False)

# Imprimir la tabla de promedio de instalaciones por género
print("Average Installs by Genre:")
print(sorted_genres)

# Recomendar el género con el mayor promedio de instalaciones
recommended_genre = sorted_genres.idxmax()
print("\nRecommended App Profile for Google Play:")
print(f"The recommended genre is '{recommended_genre}'.")

#     asume que la columna Installs ha sido limpiada y convertida a 
#     valores numéricos. La recomendación se basa en el género 
#     con el mayor promedio de instalaciones. Ten en cuenta que 
#     esta es una recomendación simplificada y hay muchos otros 
#     factores a considerar para un perfil de aplicación completo
````
